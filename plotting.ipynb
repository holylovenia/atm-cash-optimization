# Seasonality

```python
data = pd.read_csv('dataset/train/4_train.csv')
data.drop(data.columns[[1, 2, 3, 5, 6]], axis=1, inplace=True)
data.head(5)
```

```python
data['date'] = pd.to_datetime(data.date)
data = data.set_index('date')
data.head(5)
```

## Daily

```python
data['day'] = data.index.dayofweek
data.head(5)
```

```python
data[['day', 'Withdrawals']].groupby('day').mean().plot(kind='bar', legend=None)
```

```python
data[['day', 'Withdrawals']].groupby('day').mean() / data[['day', 'Withdrawals']].groupby('day').mean().max()
```

```python
data[['day', 'Withdrawals']].groupby('day').mean()
```

```python
data['month'] = data.index.month_name()
data.head(5)
```

```python
data[['month', 'Withdrawals']].groupby('month').mean().plot(kind='bar', legend=None)
```
